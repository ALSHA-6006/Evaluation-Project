# cross validation

```python
from sklearn.model_selection import cross_val_score

np.random.seed(10)
def rmse_cv(model, x,y):
    rmse =- (cross_val_score(model, x,y, scoring='neg_mean_squared_error', cv=10))
    return(rmse)

models = [LinearRegression(),
             Ridge(),
             SVR(kernel='linear'),
             SVR(kernel='poly'),
             SVR(kernel='rbf'),
             RandomForestRegressor(),
             DecisionTreeRegressor(),  
             GradientBoostingRegressor(),]

names = ['LR','R','svr','svr_p','svr_r','RF','DTR','GBR']

for model,name in zip(models,names):
    score = rmse_cv(model,x,y)
    print("{}    : {:.6f}, {:4f}".format(name,score.mean(),score.std()))
```

Accourding to all metrics score selecting the GradientBoostingRegressoras for GridSeachCV

# With Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

# Assuming your features are in `x_train` and target variable in `y_train`

# Create Gradient Boosting Regressor object
GF = GradientBoostingRegressor()

# Define parameter grid for GridSearchCV
param_grid = {
    'criterion': ['friedman_mse', 'squared_error'],  # Valid options
    'n_estimators': [100, 200]  # Reasonable range for number of trees
}

# Create GridSearchCV object with correct estimator and scoring metric
GF_grid = GridSearchCV(estimator=GF, param_grid=param_grid, cv=10, scoring='neg_mean_squared_error', n_jobs=-1, verbose=2)

# Fit the model to training data
GF_grid.fit(x_train, y_train)

# Get the best estimator (model) from GridSearchCV
best_model = GF_grid.best_estimator_

# Make predictions on test data using the best model
GF_grid_pred = best_model.predict(x_test)
```
