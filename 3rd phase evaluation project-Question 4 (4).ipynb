```python
# Step 6: Feature Selection
features = ['Latitude', 'Longitude', 'Votes', 'Aggregate rating', 'Has Table booking', 'Has Online delivery', 'Is delivering now']
X = df3[features]
```

```python
# Predicting 'Average Cost for two'
y_cost = df3['Average Cost for two']
```

```python
# Predicting 'Price range'
y_range = df3['Price range']
```

```python
# Step 7: Split the Data
X_train_cost, X_test_cost, y_train_cost, y_test_cost = train_test_split(X, y_cost, test_size=0.2, random_state=42)
X_train_range, X_test_range, y_train_range, y_test_range = train_test_split(X, y_range, test_size=0.2, random_state=42)
```

```python
# Step 8: Train Machine Learning Models
# Model for 'Average Cost for two'
regressor = RandomForestRegressor(n_estimators=100, random_state=42)
regressor.fit(X_train_cost, y_train_cost)
```

```python
# Model for 'Price range'
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train_range, y_train_range)
```

```python
# Step 9 and 10: Evaluate the Models
# Evaluate regression model
y_pred_cost = regressor.predict(X_test_cost)
rmse_cost = np.sqrt(mean_squared_error(y_test_cost, y_pred_cost))
print(f'RMSE for Average Cost for two: {rmse_cost}')
```

```python
# Evaluate classification model
y_pred_range = classifier.predict(X_test_range)
accuracy_range = accuracy_score(y_test_range, y_pred_range)
print(f'Accuracy for Price range: {accuracy_range}')
```
