```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Rainfall%20Forecast/Rainfall.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.dtypes
```

```python
import warnings
warnings.filterwarnings('ignore')
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Create the SimpleImputer object with strategy 'most_frequent'
imp = SimpleImputer(strategy='most_frequent')

# List of columns to impute
columns_to_impute = ['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

# Loop through each column and apply the imputer
for col in columns_to_impute:
    df[col] = imp.fit_transform(df[[col]])

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
list=['Date','Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','RainTomorrow']
for val in list:
    df[val]=le.fit_transform(df[val].astype(str))
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot for selected features
sns.pairplot(df[['MinTemp', 'MaxTemp', 'Rainfall', 'Humidity9am', 'Humidity3pm', 'RainTomorrow']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data for classification
X_class = df.drop(['RainTomorrow'], axis=1)
y_class = df['RainTomorrow']
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train_class, y_train_class)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Make predictions
y_pred_class = clf.predict(X_test_class)

# Evaluate the model
print(confusion_matrix(y_test_class, y_pred_class))
print(classification_report(y_test_class, y_pred_class))
```

```python
# Split the data for regression
X_reg = df.drop(['RainTomorrow'], axis=1)
y_reg = df['RainTomorrow']
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train a Random Forest Regressor
reg = RandomForestRegressor(n_estimators=100, random_state=42)
reg.fit(X_train_reg, y_train_reg)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions
y_pred_reg = reg.predict(X_test_reg)

# Evaluate the model
print(f'Mean Squared Error: {mean_squared_error(y_test_reg, y_pred_reg)}')
print(f'R2 Score: {r2_score(y_test_reg, y_pred_reg)}')
```
