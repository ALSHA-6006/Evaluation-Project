```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
list=['Date']
for column in list:
    df[column]=le.fit_transform(df[column].astype(str))
```

```python
import sklearn
from sklearn.impute import SimpleImputer
list=['station','Present_Tmax','Present_Tmin','LDAPS_RHmin','LDAPS_RHmax','LDAPS_Tmax_lapse','LDAPS_Tmin_lapse','LDAPS_WS','LDAPS_LH','LDAPS_CC1','LDAPS_CC2','LDAPS_CC3','LDAPS_CC4','LDAPS_PPT1','LDAPS_PPT2','LDAPS_PPT3','LDAPS_PPT4','Next_Tmax','Next_Tmin']
imp = SimpleImputer(strategy='mean')
for column in list:
    df[[column]] = imp.fit_transform(df[[column]])
```

# Splitting the Data

```python
from sklearn.model_selection import train_test_split

# Features and targets for Next_Tmax prediction
X_tmax = df.drop(['Date', 'Next_Tmax', 'Next_Tmin'], axis=1)
y_tmax = df['Next_Tmax']

# Features and targets for Next_Tmin prediction
X_tmin = df.drop(['Date', 'Next_Tmax', 'Next_Tmin'], axis=1)
y_tmin = df['Next_Tmin']

# Split the data into training and test sets for Next_Tmax
X_tmax_train, X_tmax_test, y_tmax_train, y_tmax_test = train_test_split(X_tmax, y_tmax, test_size=0.2, random_state=42)

# Split the data into training and test sets for Next_Tmin
X_tmin_train, X_tmin_test, y_tmin_train, y_tmin_test = train_test_split(X_tmin, y_tmin, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Model for Next_Tmax
model_tmax = LinearRegression()
model_tmax.fit(X_tmax_train, y_tmax_train)
```

```python
# Model for Next_Tmin
model_tmin = LinearRegression()
model_tmin.fit(X_tmin_train, y_tmin_train)
```

```python
# Predictions for Next_Tmax
y_tmax_pred = model_tmax.predict(X_tmax_test)
print("Next_Tmax Model Evaluation:")
print("MAE:", mean_absolute_error(y_tmax_test, y_tmax_pred))
print("MSE:", mean_squared_error(y_tmax_test, y_tmax_pred))
print("R2 Score:", r2_score(y_tmax_test, y_tmax_pred))
```

```python
# Predictions for Next_Tmin
y_tmin_pred = model_tmin.predict(X_tmin_test)
print("Next_Tmin Model Evaluation:")
print("MAE:", mean_absolute_error(y_tmin_test, y_tmin_pred))
print("MSE:", mean_squared_error(y_tmin_test, y_tmin_pred))
print("R2 Score:", r2_score(y_tmin_test, y_tmin_pred))
```
